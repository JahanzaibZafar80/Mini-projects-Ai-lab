```python
import pandas as pd

df=pd.read_csv("car_crashes.csv")
# df.info()
df.columns
```

```python
df.info()
```

```python
def float_to_int(lis):
    for i in lis:
        df[i]=df[i].astype("int64")
lis1=['total', 'speeding', 'alcohol', 'not_distracted', 'no_previous',
       'ins_premium', 'ins_losses']
float_to_int(lis1)
```

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np
def convert_obj_int(lis):
    for i in lis:
        data=pd.DataFrame({i:df[i].unique()})
        data_le=LabelEncoder()
        data_le.fit(np.ravel(data))
        df[i]=data_le.transform(df[i])
lis1=["abbrev"]
convert_obj_int(lis1)
```

```python
df.info()
```

```python
X=df.drop("total",axis=1)
Y=df["total"]
```

# Train Test Splitting

```python
from sklearn.model_selection import train_test_split
train_X,test_X,train_Y,test_Y=train_test_split(X,Y,test_size=0.3,shuffle=False)
```

```python
print(train_X.shape,train_Y.shape)
print(test_X.shape,test_Y.shape)
```

# Model Training

```python
from sklearn.svm import SVC
model_svc=SVC()
model_svc.fit(train_X, train_Y)
print(model_svc)
```

# Prediction

```python
model_pred_svc = model_svc.predict(test_X)
model_pred_svc
```

# Accuracy

```python
from sklearn.metrics import accuracy_score
model_acc_svc = accuracy_score(test_Y, model_pred_svc)
print(round(model_acc_svc,3)*100, "%")
```

# Model Saving

```python
import pickle
pickle.dump(model_svc, open('model_svc.pkl', 'wb'))   
```
